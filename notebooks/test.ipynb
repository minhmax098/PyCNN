```python
# Read npy files
import numpy as np
```

```python
# the array is saved in the file geekfile.npy 
image = np.load('../images/images_0.npy')
```

```python
image.shape
```

```python
import matplotlib.pyplot as plt
plt.imshow(image, cmap='gray')
plt.show()
```

```python
# Save the image back as gray scale image
plt.imsave('output_image.png', image, cmap='gray')
```

```python
from pycnn import PyCNN

# Initialize object
cnn = PyCNN()
```

```python


# Perform respective image processing techniques on the given image

print("Processing images...")
cnn.edgeDetection('output_image.png', 'output1.png')
cnn.grayScaleEdgeDetection('output_image.png', 'output2.png')
cnn.cornerDetection('output_image.png', 'output3.png')
cnn.diagonalLineDetection('output_image.png', 'output4.png')
cnn.inversion('output_image.png', 'output5.png')
cnn.optimalEdgeDetection('output_image.png', 'output6.png')
```
